```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.model_selection import train_test_split
import sklearn.datasets
```

```python
df=sklearn.datasets.load_boston()
```

```python
print(df)
```

```python
df1=pd.DataFrame(df.data,columns=df.feature_names)
df1.head()
```

```python
#Adding the price column to the dataframe
df1['price']=df.target
df1
```

```python
#checking missing values
df1.isnull().sum()
```

# Splitting the data and the target

```python
x=df1.drop(columns='price',axis=1)
x
```

```python
y=df1["price"]
y
```

# Training and testing data splitting

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

# Model Training

#XGBoost Regressor

```python
from xgboost import XGBRegressor
```

```python
model=XGBRegressor()
model.fit(x_train,y_train)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
training_data_prediction=model.predict(x_train)

print(training_data_prediction)
```

```python
#R-Squared Error
Score=metrics.r2_score(y_train,training_data_prediction)
```

```python
#Mean Absolute Error
score1=metrics.mean_absolute_error(y_train,training_data_prediction)
```

```python
print("R-Squared Error:",Score)
print("Mean-Absolute-Error",score1)
```

# Visualize the actual prices on the predicted prices

```python
plt.scatter(y_train,training_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted prices")
plt.title("Actual price vs Predicted price")
plt.show()
```

# Evaluating the test data

```python
testing_data_prediction=model.predict(x_test)

print(testing_data_prediction)
```

```python
#R-Squared Error
Score=metrics.r2_score(y_test,testing_data_prediction)
#Mean Absolute Error
score1=metrics.mean_absolute_error(y_test,testing_data_prediction)
print("R-Squared Error:",Score)
print("Mean-Absolute-Error",score1)
```
